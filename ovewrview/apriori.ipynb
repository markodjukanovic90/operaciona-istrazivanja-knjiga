```python
import numpy as np 
import pandas as pd
from apyori import apriori  
```

```python
data = pd.read_csv("/content/Market_basket_data.csv")
data

transacitions = []
for i in range(data.shape[0]):
  items_person = [ str(data.iloc[i, j]) for j in range(  data.shape[ 1 ]) if data.iloc[i, j] is not np.nan  ]
  transacitions.append( items_person )

transacitions

rules = apriori(transactions=transacitions, min_support = 0.03, min_confidence = 0.2, min_lift = 0.1, min_length = 2, max_length = 2)
res = list(rules) 

print("Sljedece zavisnosti su primijecene: ")
for item in res:
  pair = item[ 0 ]
  item = [ x for x in pair ]
  if len(item) >= 2:
    print( item[ 0 ], "===> ", item[ 1 ]) 

# min_support ==> A => B se pojavljuje u 3% posto slucajeva 
# min_confidence ==> znaci da u barem 20% slucajeva ljudi koji kupe namirnicu A, kupe i namirnicu B...
```
